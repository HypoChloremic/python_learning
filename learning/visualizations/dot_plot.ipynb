# Dot plotting
There are several parameters that need to be considered when performing a dotplot. 
For the data, several aspects needs to be considered.

For the dot plot, each term, needs to be represented on the entire axis, for each tick. Say for the $y$-axis 10 labels are present, and for the $x$-axis 3 labels are present, then the matrix needs to be a $10\times 3$ matrix:
$$
\text{the data values}: \\
x-axis = \langle x_1, x_2 . . . \rangle \\
y-axis = \langle y_1, y_2 . . . \rangle \\
m-matrix = 
\begin{bmatrix}
a_{11}, a_{12}, . . . \\ 
\vdots, \ddots, \vdots \\
a_{n1}, a_{n2}, ...
\end{bmatrix} \\
$$

```python
up = 10
dw = 0
ln = up-dw + 1
xs = np.linspace(dw, up, ln)
ys = np.linspace(dw, up, ln)
grid = np.array(np.meshgrid(np.array(xs), np.array(ys)))
grid_ = np.stack(np.concatenate(np.stack(grid,axis=-1)), axis=-1)
names = [i for l in 'abcdefghijk' for i in [l]*11]
```

### Colors

```python
alphas = np.linspace(0.1, 1, 10)
rgba_colors = np.zeros((121,4))
# for red the first column needs to be one
rgba_colors[:,0] = 1.0
# the fourth column needs to be your alphas
rgba_colors[:, 3] = [i/121 for i in range(121)]
rgba_colors
```

```python

fig, ax = plt.subplots()
scatter = ax.scatter(grid_[0], names, s=[np.abs(20*np.sin(i)) for i in range(121)], color=rgba_colors)


# produce a legend with a cross section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, title="Sizes", bbox_to_anchor=(2, 0.5))
```
