```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
url = 'https://python-graph-gallery.com/wp-content/uploads/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
df
```

```python
dendrogram(linkage(df.T), labels=df.columns)
```

### Based on example
What can be seen is that the values of each vector is taken into account. The dendrogram then sorts them based on the index of the vector in the matrix

```python
l = linkage(np.array([[1,2], [1,2], [3,4]]))
print(l)
dendrogram(l)
```
